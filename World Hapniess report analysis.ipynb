## 8. Indentify the coorelation between six variables and index score

```python
# Creating a numerical data for calculating the coorelation
numericalData = data19[['Score', 'GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices']]

numericalData['Generosity'] = data19['Generosity']
numericalData['Perceptions of corruption'] = data19['Perceptions of corruption']
```

```python
plt.figure(figsize=(34,20))
plt.title(label='Coorelation between Happiness Score and other variables')
sns.heatmap(data=numericalData.corr(), annot=True, cmap='magma', linecolor='white', linewidths=4)
```

1. The above heatmap makes it very clear that Happiness score is most influenced by three variables. I.e GDP per capita, Social support and Healthy Life expentancy.

2. It really makes sense because High GDP per capita gives more purchasing power, Social support is needed for people to feel good and long life expectancy means less diseases, less likely to die young and good healtcare present in the country.

## 9. Further Analysis of variable with happiness score

```python
plt.figure(figsize=(12,6))
plt.title(label='GDP per Capita vs Happniess Score')
sns.scatterplot(data=data19, x='GDP per capita', y='Score', hue='Score');
```

As we saw in the heatmap that there is a positive correlation between GDP per capita and Hapiness Score. The higher the GDP per capita the better the happiness score is

```python
plt.figure(figsize=(12,6))
plt.title(label='Social support vs Happniess Score')
sns.scatterplot(data=data19, x='Social support', y='Score', hue='Score');
```

As we saw in the heatmap that there is a positive correlation between Social Support and Hapiness Score. The higher the social support the better the happiness score is

```python
plt.figure(figsize=(12,6))
plt.title(label='Healthy life expectancty vs Happniess Score')
sns.scatterplot(data=data19, x='Healthy life expectancy', y='Score', hue='Score');
```

As we saw in the heatmap that there is a positive correlation between Life expectancy and Hapiness Score. The higher the social support the better the happiness score is

```python
plt.figure(figsize=(12,6))
plt.title(label='Freedom to make life choices vs Happniess Score')
sns.scatterplot(data=data19, x='Freedom to make life choices', y='Score', hue='Score');
```

Freedom to make life choices is high in the countries who score really good on happiness score. Also score and freedom to make life choices are moderatly correlated as shown in the heatmap.
